```python
# delete the Flight# column as it is not necessary for the aim of this project
airplane_data = airplane_data.drop('Flight #', axis=1)
```

```python
airplane_data
```

```python
# delete the Registration column as it is not necessary for the aim of this project
airplane_data = airplane_data.drop('Registration', axis=1)
```

```python
airplane_data
```

```python
# airplane_data['Time'] = airplane_data['Time'].fillna('00:00')
```

```python
airplane_data['Time'].isnull().sum()
```

```python
# fill up empty cells within each of these categorical columns
airplane_data['Route'] = airplane_data['Route'].fillna('Not Defined')
airplane_data['Summary'] = airplane_data['Summary'].fillna('No Comments')
airplane_data['Operator'] = airplane_data['Operator'].fillna('Unknown')
airplane_data['Type'] = airplane_data['Type'].fillna('No Type')
```

```python
airplane_data.isnull().sum()
```

```python
# since we want to check whether or not an operator is military or not, we would check each of the cells for any occurences
# of the string 'military' and print either true or flase
airplane_data['Is Military'] = airplane_data['Operator'].str.contains('Military', regex=False)
```

```python
airplane_data['Is Military'].value_counts()
```

```python
airplane_data.head()
```

```python
# the location column contains states and country in some instances and only the country is needed.
# slipt the values of each cell by the ',' delimeter, then extract the lastt string and output it in uppercase
airplane_data['Location Country']  = airplane_data['Location'].str.split(',').str[-1].str.strip().str.upper()
```

```python
airplane_data['Location Country']
```

```python
airplane_data['Survived'] = airplane_data['Aboard'] - airplane_data['Fatalities']
```

```python
airplane_data
```

```python
airplane_data['Survival Rate'] = (airplane_data['Survived']/airplane_data['Aboard']) * 100
```

```python
airplane_data
```

```python
import random
list_country = airplane_data['Location Country'].to_list()
pick_country = random.choice(list_country)
```
